```python
import numpy as np 
import librosa
import librosa.display
import IPython.display as ipd
from scipy.io.wavfile import read
import matplotlib.pyplot as plt 
```

```python
wavPath = "happy.wav"
x, fs = librosa.load(wavPath)
sr, _ = read(wavPath)
t = np.arange(0,len(x)/sr, 1./sr)
plt.plot(t,x[:len(t)])
```

```python
ipd.Audio(wavPath)
```

```python
# 16000 --> 1000
# fsize --> 32

frame_size = 512
hop_size = 256 
nbFrames = int((len(x)-frame_size)/hop_size) + 1
rms = []
for i in range(nbFrames):
    frame_i = x[i*hop_size:i*hop_size + frame_size]
    s=0
    for j in range(len(frame_i)):
        s+=frame_i[j]**2/frame_size
    rms.append(np.sqrt(s))
```

```python
t = librosa.frames_to_time(np.arange(nbFrames), sr=sr, hop_length=hop_size,)
librosa.display.waveshow(x, sr=sr)
plt.plot(t, rms)
```
